## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
Decision: Which specific pages in our content library should the content team invest time and budget into updating today?

Unit of Analysis: A unique page (content_hash_id or url_hash_id).

Output: A ranked review queue sorted by a priority score (0–100) along with actionable reason codes (e.g., stale_high_traffic, thin_content_decline).

Action: A content writer or editor pulls the top pages from the queue, inspects them, and executes a targeted content refresh or optimization.

Cost of a Wrong Call:

False Positive (Low cost): A writer spends an hour updating a page that didn't strictly need it.

False Negative (High cost): A highly valuable page undergoing a quiet traffic drain is ignored, leading to compounding, long-term revenue loss.

Justification for Data/ML: Instead of maintaining rigid, arbitrary heuristic rules (like "refresh every page older than 6 months") which fail to scale and ignore actual traffic dynamics, ML can learn complex patterns of decline and recovery across thousands of pages to forecast the true potential traffic lift.
```

## 3. Quick look at the data (2-3 real numbers)

Percentage of pages actively declining: 54.21%
Average baseline impressions per page: 5200.37 weeks.*

```python
import pandas as pd

# Load the dataset directly from your internet repository
url = "https://raw.githubusercontent.com/ahmedumer1941/Flyrank-Internship/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# 1. Total volume of pages in our dataset
total_pages = len(df)
print(f"Total pages in starter dataset: {total_pages}")

# 2. Proportion of pages currently experiencing a downward trend
declining_pct = (df['trend_direction'] == 'down').mean() * 100
print(f"Percentage of pages actively declining: {declining_pct:.2f}%")

# 3. Baseline traffic scale (fixed column name)
avg_impressions = df['impressions_90d'].mean()
print(f"Average baseline impressions per page: {avg_impressions:.2f}")
```

What I can claim: This model will provide decision-support to help prioritize human attention. It surfaces observed historical correlations and directional flags indicating which pages match known risk profiles of content decay.

What I cannot claim: This model does not establish causal proof (e.g., it cannot guarantee that updating a page will definitely reverse a traffic drop). It is absolutely not "predicting Google's core algorithm" or forecasting exact future search engine result page (SERP) ranks.
