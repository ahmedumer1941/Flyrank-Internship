## 1. Unit of analysis + time window

The unit of analysis (grain) for this data contract is a unique combination of content identifier (`content_hash_id`) and client domain (`client_hash_id`) aggregated over a single date (`report_date`). The target exploration window is the mid-panel month of March 2026 (from 2026-03-01 to 2026-03-31).

```python
import pandas as pd
from datasets import load_dataset

# 1. Directly target only the March 2026 data slice from the repository
print("Loading March 2026 slice directly...")
march_dataset = load_dataset(
    'FlyRank/internship-warehouse',
    'fact_content_daily_performance',
    data_files="fact_content_daily_performance/month=2026-03/*.parquet",
    split='train'
)

# 2. Instantly convert this small, clean slice into a pandas DataFrame
df_march = pd.DataFrame(march_dataset)

# 3. Print out your validation numbers
print(f"\nTotal rows in March 2026: {len(df_march)}")
print(f"Date range: {df_march['report_date'].min()} to {df_march['report_date'].max()}")

# 4. View a preview of your active columns
df_march[['report_date', 'client_hash_id', 'content_hash_id', 'gsc_impressions', 'ga4_sessions']].head()
```

## 3. Verify it with queries (grain, counts, missing values, windows)

### Verification & Quality Gate Metrics
* **Grain Verification (Query 1):** We verify that the grain is structurally unique across `report_date`, `client_hash_id`, and `content_hash_id` by asserting that the total row count perfectly matches the count of unique grouping combinations.
* **Temporal Integrity (Query 2):** The dataset is explicitly bounded inside our target month window spanning from 2026-03-01 to 2026-03-31 with a sub-slice total of 9,841,378 active event records.
* **Data Availability (Query 3):** We measure active platform integration data quality using the `ga4_data_available` gate to track the survival baseline rate of valid user traffic telemetry metrics.

```python
# --- QUERY 1: Verify the Grain (Uniqueness Check) ---
total_rows = len(df_march)
unique_grain_combinations = len(df_march.groupby(['report_date', 'client_hash_id', 'content_hash_id']))

print(f"Query 1 - Grain Verification:")
print(f"  > Total Row Count: {total_rows}")
print(f"  > Unique [date + client + content] Combinations: {unique_grain_combinations}")
print(f"  > Grain is Unique: {total_rows == unique_grain_combinations}\n")


# --- QUERY 2: Row Count & Temporal Date Span ---
min_date = df_march['report_date'].min()
max_date = df_march['report_date'].max()

print(f"Query 2 - Temporal Boundary Check:")
print(f"  > Row count for targeted slice: {total_rows}")
print(f"  > Exact operational window: {min_date} to {max_date}\n")


# --- QUERY 3: Availability Check (Surviving Quality Gate) ---
# Check how many rows have valid active GA4 metrics tracking data
surviving_rows = df_march[df_march['ga4_data_available'] == True]
survival_count = len(surviving_rows)

print(f"Query 3 - Availability Gate Check:")
print(f"  > Rows with ga4_data_available == True: {survival_count}")
print(f"  > Survival Rate: {(survival_count / total_rows) * 100:.2f}%")
```

```python
# Create the clean 5-feature training frame matrix from March data
feature_matrix = pd.DataFrame()

# 1. Context fields (Identifier Keys)
feature_matrix['report_date'] = df_march['report_date']
feature_matrix['client_hash_id'] = df_march['client_hash_id']
feature_matrix['content_hash_id'] = df_march['content_hash_id']

# 2. Five distinct behavioral features (Knowable at the decision moment)
feature_matrix['feat_gsc_impressions'] = df_march['gsc_impressions'].fillna(0)
feature_matrix['feat_gsc_avg_position'] = df_march['gsc_avg_position'].fillna(0)
feature_matrix['feat_ga4_pageviews_baseline'] = df_march['ga4_pageviews'].fillna(0)
feature_matrix['feat_scroll_events_baseline'] = df_march['scroll_events'].fillna(0)
feature_matrix['feat_organic_ratio'] = (df_march['sessions_organic'] / (df_march['ga4_sessions'] + 1e-5)).fillna(0)

# 3. Target label proxy field
feature_matrix['label_high_traffic'] = (df_march['ga4_sessions'] > 10).astype(int)

print(f"Feature matrix built successfully with {feature_matrix.shape[1]} columns!")
feature_matrix.head()
```

## 4. Data limits

This dataset cannot capture intra-day search trend fluctuations or immediate conversion attribution because it is aggregated to a strict daily granularity. Furthermore, historical baselines for newly created URLs are structurally missing (causing cold-start blind spots), and because the pipeline excludes platform operational flags (`client_has_ga4`, etc.), the model cannot inherently tell if a sudden drop in metrics is a true behavioral loss or a technical tracking outage.

```python
# Check what percentage of our features default to zero (data gaps/limits)
missing_baseline_pct = (df_march['ga4_pageviews'].isna().sum() / len(df_march)) * 100
print(f"Percentage of rows with missing/null historical pageviews: {missing_baseline_pct:.2f}%")

# Double-check data range integrity to ensure no leaking boundaries
print(f"Minimum date in window: {df_march['report_date'].min()}")
print(f"Maximum date in window: {df_march['report_date'].max()}")
```
